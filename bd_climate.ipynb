# 1. In which years were the highest and lowest temperatures recorded? #

```python
import pandas as pd
df = pd.read_csv('sorted_temp_and_rain_dataset.csv')
max_em_rowt = df.loc[df['tem'].idxmax()]
min_tem_row = df.loc[df['tem'].idxmin()]
print('max tem:',max_tem_row['tem'],"c - year:",max_tem_row['Year'],'Month',max_tem_row['Month'])
print('max tem:',min_tem_row['tem'],"c - year:",min_tem_row['Year'],'Month',min_tem_row['Month'])
```

# 2.How has the average annual temperature changed over the past 50 years? #

```python
annual_tem = df.groupby('Year')['tem'].mean().reset_index()
annual_tem 
```

```python
last_50_years = annual_tem.tail(50)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(last_50_years['Year'],last_50_years['tem'],marker = '*',color = 'orange')
plt.title('last_50_years_ave_tem')
plt.xlabel('Year')
plt.ylabel('ave_tem')
plt.grid(True)
```

# 3.What are the seasonal patterns of rainfall? In which months does it usually rain the most? #

```python
monthly_rain = df.groupby('Month')['rain'].mean()
monthly_rain.plot(kind = 'bar',color = 'skyblue',figsize = (10,6))
plt.title('ave_rain')
plt.xlabel('Month')
plt.ylabel('ave_rainfall (mm)')
plt.xticks(ticks = range(0,12),labels = ['jan','feb','mar','april','may','june','july','aug','sep','oct','nov','dec'],rotation = 45)
plt.grid(axis = 'y')
plt.show()
```

# 4.Is there any correlation between temperature and rainfall? #

```python
corr = df['tem'].corr(df['rain'])
print('tem_rainfall_corr :',corr)
```
